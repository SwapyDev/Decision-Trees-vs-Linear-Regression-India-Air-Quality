```python
#Create a cleaned model variable to drop the date because we'll use it for analysis later
train_df = df.drop(['City', 'Date', 'NO', 'NOx', 'Benzene', 'Toluene', 'Xylene', 'AQI_Bucket'], axis=1)
```

```python
#Dataset for training
train_df.head()
```

```python
#Initialize x_train
x = train_df[["PM2.5", "PM10", "NO2", 'NH3', 'CO','SO2','O3']]
x
```

```python
#Initialize y
y = train_df[['AQI']]
y
```

```python
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2 )
```

```python
#Scale the data
scaler = StandardScaler().fit(x_train, x_test)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#Initialize linear regression model
linear_reg = LinearRegression()
```

```python
model = linear_reg.fit(x_train, y_train)
yhat = model.predict(x_test)
accuracy = model.score(x_train, y_train)
print(f'Prediction: {accuracy: .4f}')
```

```python
#Initialize decision tree 
from sklearn.tree import DecisionTreeRegressor
decision_tree = DecisionTreeRegressor()
```

```python
model = decision_tree.fit(x_train, y_train)
y_hat = model.predict(x_test)
accuracy = model.score(x_train, y_train)
print(f'Score: {accuracy: 4f}')
```
